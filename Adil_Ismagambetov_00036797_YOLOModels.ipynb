```python
import glob
import numpy as np

test_label_path = "/content/dataset/dataset/labels/test/*.txt"

true_labels = []
true_boxes = []
for label_file in glob.glob(test_label_path):
    with open(label_file) as f:
        for line in f.readlines():
            class_id = int(line.strip().split()[0])
            true_labels.append(class_id)
            true_boxes.append([*map(float, line.strip().split()[1:])])

true_labels = np.array(true_labels)
```

```python
from pathlib import Path
import torch

pred_files = sorted(Path("/content/yolov5/runs/val/unweighted_model_results3/labels").glob("*.txt"))

pred_labels = []
pred_boxes = []
confidences = []
for pred_file in pred_files:
    with open(pred_file) as f:
        for line in f.readlines():
            parts = line.strip().split()
            if len(parts) >= 6:  # Format: class x1 y1 x2 y2 conf
                pred_labels.append(int(parts[0]))
                pred_boxes.append([*map(float, parts[1:5])])
                confidences.append(float(parts[5]))

pred_labels = np.array(pred_labels)
confidences = np.array(confidences)
```

```python
print(f"Ground truth samples: {len(true_labels)}")
print(f"Predicted samples: {len(pred_labels)}")
print("Class distribution (True vs Pred):")
print(np.unique(true_labels, return_counts=True))
print(np.unique(pred_labels, return_counts=True))
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

precision = precision_score(true_labels, pred_labels, average='weighted')
recall = recall_score(true_labels, pred_labels, average='weighted')
f1 = f1_score(true_labels, pred_labels, average='weighted')

print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-score: {f1:.4f}")
```

```python
import torch
from torchmetrics.detection import MeanAveragePrecision


def yolo_to_torchmetrics(pred_files, gt_files):
    preds, targets = [], []

    for pred_file, gt_file in zip(sorted(pred_files), sorted(gt_files)):
        # Load predictions
        pred_data = np.loadtxt(pred_file)
        if len(pred_data) > 0:
            pred_data = pred_data.reshape(-1, 6) if pred_data.ndim == 1 else pred_data
            preds.append({
                'boxes': torch.tensor(pred_data[:, 1:5]),  # x1,y1,x2,y2
                'scores': torch.tensor(pred_data[:, 5]),  # conf
                'labels': torch.tensor(pred_data[:, 0].astype(int)),  # class
            })

        # Load ground truth
        gt_data = np.loadtxt(gt_file)
        if len(gt_data) > 0:
            gt_data = gt_data.reshape(-1, 5) if gt_data.ndim == 1 else gt_data
            targets.append({
                'boxes': torch.tensor(gt_data[:, 1:5]),
                'labels': torch.tensor(gt_data[:, 0].astype(int)),
            })

    return preds, targets

# Usage
pred_files = list(Path("/content/yolov5/runs/val/unweighted_model_results3/labels").glob("*.txt"))
gt_files = list(Path("/content/dataset/dataset/labels/test").glob("*.txt"))
preds, targets = yolo_to_torchmetrics(pred_files, gt_files)
```

```python
metric = MeanAveragePrecision(
    box_format='xyxy',
    iou_thresholds=[0.5, 0.75],  # Standard COCO thresholds
    class_metrics=True  # Enable per-class metrics
)

metric.update(preds, targets)
results = metric.compute()

print(f"mAP for YOLOv5 Unweighted Model: {results['map'].item():.4f}")
```

mAP is worse for YOLOv5 Unweighted Model, but it trains much faster
